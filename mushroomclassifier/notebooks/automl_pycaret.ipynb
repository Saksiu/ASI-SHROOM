```python
### Mushroom AutoML with PyCaret


# 1. Importy
import pandas as pd
from pycaret.classification import *
import joblib

# 2. Wczytaj dane
DATA_PATH = "../data/01_raw/mushrooms.csv"
df = pd.read_csv(DATA_PATH)

# 3. Dodaj nagłówki kolumn (zgodnie z dokumentacją UCI)
df.columns = [
    "class", "cap-shape", "cap-surface", "cap-color", "bruises", "odor",
    "gill-attachment", "gill-spacing", "gill-size", "gill-color",
    "stalk-shape", "stalk-root", "stalk-surface-above-ring", "stalk-surface-below-ring",
    "stalk-color-above-ring", "stalk-color-below-ring", "veil-type", "veil-color",
    "ring-number", "ring-type", "spore-print-color", "population", "habitat"
]

# 4. Uruchom PyCaret Setup
clf = setup(
    data=df,
    target="class",
    session_id=42,
    verbose=True,
    normalize=False
)

# 5. Porównaj modele
best_model = compare_models()

# 6. Stwórz cały pipeline (model + preprocessing)
best_model_final = finalize_model(best_model)

# 7. Zapisz model
joblib.dump(best_model_final, "../data/06_models/best_model.pkl")

# 8. (opcjonalnie) Wyświetl ranking
models = pull()
models.head(10)

```
